have the units of cubic meter per capita per day.

## Clustering Analysis

The purpose of clustering analysis is:
- to detect data point outliers.
- to decompose *water consumption* into *base consumption* and *seasonal consumption*.

The attributes used in the clustering are *water consumption*, *minimum temperature* and *maximum temperature*. Through the clustering analysis, hidden groups in data points are found and the influence of seasonality (air temperature) on *water consumption* is verified.

Clustering analysis fails in handling missing data. Before performing the clustering analysis, the missing values are replaced by interpolation.

```python
#Select water consumption, minimum temperature and maximum temperature
df_cluster = df.loc[:,['UWC', 'Min Temp (C)', 'Max Temp (C)']]

#Replace missing values
df_cluster.interpolate(inplace = True)

#Data normalization, Z transformation
df_cluster = (df_cluster - df_cluster.mean()) / df_cluster.std()
```

### Dendrogram
Dendrogram showes the Euclidean distances for possible data clusters.

```python
Z = linkage(df_cluster, 'ward')
dendrogram(Z, truncate_mode = 'lastp', p=12, leaf_rotation = 45., leaf_font_size = 12., show_contracted = True)

plt.title('Dendrogram',  {'fontsize': 15})
plt.xlabel('Cluster size', {'fontsize': 12})
plt.ylabel('Distance', {'fontsize': 12})
plt.subplots_adjust(bottom = 0.2)
```

### Hierarchical Clustering by Agglomerative Nesting

We use a distance matrix as clustering criteria and implement hierarchical clustering by agglomerative nesting. This algorithm begins with assigning each data point to a cluster by its own. Therefore, initially, the total number of clusters is equal to the number of data points. Then, it merges the data points of the least dissimilarity until all the data points belong to the same cluster. 

In this research, the agglomerative nesting is based on Euclidian distance, which is a measure of similarity levels. The performance of the clustering algorithm is evaluated by using the Silhouette coefficient.

**Note:** Click [here](https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html) to learn more about Silhouette analysis.

```python
# Agglomerative (Hierarchiacal Clustering)

range_n_clusters = [2, 3, 4]
for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 3 columns
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
    fig.set_size_inches(18, 7)
    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(df_cluster) + (n_clusters + 1) * 10])
    clusterer = AgglomerativeClustering(affinity = 'euclidean', compute_full_tree = 'auto',
                                        connectivity = None, linkage = 'ward', memory = None, 
                                        n_clusters = n_clusters, pooling_func = 'deprecated')
    cluster_labels = clusterer.fit_predict(df_cluster)

    silhouette_avg = silhouette_score(df_cluster, cluster_labels)
    print("For n_clusters =", n_clusters,"The average silhouette_score is :", silhouette_avg)

     # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(df_cluster, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor = color, edgecolor = color, alpha = 0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples
        
    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")


    # The vertical line for average silhouette score of all the values
    ax1.axvline(x = silhouette_avg, color = "red", linestyle = "--")
    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing water consumtion time series
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(y = df['UWC'], x = df['Date'], marker='.', s = 30, lw = 0, alpha = 0.7,
                c = colors, edgecolor = 'k')
    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Date")
    ax2.set_ylabel("UWC ($m^3$/capita.day)")
    plt.suptitle(("Silhouette analysis for Hierarchical clustering "
                  "for n_clusters = %d" % n_clusters),
                 fontsize = 14, fontweight = 'bold')


    # 3rd Plot showing water consumption versus mean temperature
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax3.scatter(y = df['UWC'], x = df['Mean Temp (C)'], marker = '.', s = 30, lw = 0, 
                alpha = 0.7, c = colors, edgecolor = 'k')
    ax3.set_title("The visualization of the clustered data.")
    ax3.set_xlabel("Mean temperature (C)")
    ax3.set_ylabel("UWC ($m^3$/capita.day)")
    plt.suptitle(("Silhouette analysis for Hierarchical clustering "
                  "for n_clusters = %d" % n_clusters),
                 fontsize = 14, fontweight = 'bold')


clusterer = AgglomerativeClustering(affinity = 'euclidean', compute_full_tree = 'auto',
                                        connectivity = None, linkage='ward', memory = None, 
                                        n_clusters = 3, pooling_func = 'deprecated')
clusterer.fit(df_cluster)
# Creat a column to show cluster labels.
df['Cluster'] = clusterer.labels_
```

## Detection and Replacement of Data Point Outliers
Two methods are used to implement outlier detection:
- Clustering-based method 
- Statistical methods 

Both the clustering-based and statistical methods are used to detect unusual patterns and outliers. The statistical methods assume a normal distribution of data points. Therefore, values in a low probability region are considered as outliers. The clustering-based methods accept that outliers might belong to a small or sparse cluster or might be far from the clusters to which they are closest. 

In this research, upon detecting outliers, they are replaced by the average between the two values for the dates that precede and follow the date of the outlier.

```python
# Define lower bound and upper bound of high probability region
pd.options.mode.chained_assignment = None

up_b = df_b['UWC'].mean() + 2*df_b['UWC'].std()
low_b = df_b['UWC'].mean() - 2*df_b['UWC'].std()

up_s = df_s['UWC'].mean() + 2*df_s['UWC'].std()
low_s = df_s['UWC'].mean() - 2*df_s['UWC'].std()

# Replace outlier by interpolation for base consumption
df_b.loc[df_b['UWC'] > up_b, 'UWC'] = np.nan
df_b.loc[df_b['UWC'] < low_b, 'UWC'] = np.nan
df_b['UWC'].interpolate(inplace=True)                          

# Replace outlier by interpolation for seasonal consumption
df_s.loc[df_s['UWC'] > up_s, 'UWC'] = np.nan
df_s.loc[df_s['UWC'] < low_s, 'UWC'] = np.nan
df_s['UWC'].interpolate(inplace=True)                            

df_s['Max Temp (C)'].interpolate(inplace = True)  
df_b['Max Temp (C)'].interpolate(inplace = True)  

df_s['Min Temp (C)'].interpolate(inplace = True)  
df_b['Min Temp (C)'].interpolate(inplace = True)
```

## Temperature Threshold
The transition from *base consumption* to *seasonal consumption* occurs at temperature threshold. We find temperature threshold from scatter plot and the best fit of *water consumption* versus *mean temperature*. 

```python
fig = plt.figure()
fig.set_size_inches(8, 6)

# Define a function to plot best fits for 2012, 2013, 2014 and 2012 - 2014
def Bestfit(StartD, EndD, color, ls, year):
    
    # Filter a dataset for base consumption and seasonal consumption for a selected time spane
    time_s = (df_s['Date'] > StartD) & (df_s['Date'] <= EndD)
    dataframe_s = df_s.loc[time_s]
    
    time_b = (df_b['Date'] > StartD) & (df_b['Date'] <= EndD)
    dataframe_b = df_b.loc[time_b]
    
    # Define x-axis and y-axis values
    X_s = dataframe_s.loc[:,['Mean Temp (C)']]
    Y_s = dataframe_s.loc[:,['UWC']]
    res_s = stats.theilslopes(Y_s, X_s)
    
    X_b = dataframe_b.loc[:,['Mean Temp (C)']]
    Y_b = dataframe_b.loc[:,['UWC']]
    res_b = stats.theilslopes(Y_b, X_b)
    b = -((res_b[1]-res_s[1])/(res_b[0]-res_s[0]))
    
    # Plot lines of best fits
    x_b = np.linspace(-20,b)
    y_b = res_b[0]* x_b + res_b[1]
    x_s = np.linspace(b, 25)
    y_s = res_s[0]* x_s + res_s[1]

    plt.plot(x_s, y_s, color = color, ls = ls , label=year)
    plt.plot(x_b, y_b, color = color,  ls = ls)
    plt.legend()
    plt.grid()
    plt.xlabel('Mean Temp (C)')
    plt.ylabel('UWC ($m^3$/capita.day)')
    plt.title('Best fit')
    print("Threshold in " + year + ":", b)


    
Bestfit('2011-12-31', '2012-12-31', 'black', "--", '2012')
Bestfit('2012-12-31', '2013-12-31', 'black',"-.", '2013')
Bestfit('2013-12-31', '2014-12-31', 'black', ":",'2014')
Bestfit('2011-12-31', '2014-12-31', 'black', "-",'2012-2014')
plt.grid()

plt.show()
```

# Bayesian Linear Regression
Bayesian statistics is a powerful probabilistic modeling that we use for *seasonal consumption* forecasting. *Seasonal consumption* is a function of *maximum temperature*, *minimum temperature* and *total precipitation*. 



```python
# Define seasonal consumption dataset
df_s = df_s.loc[:, ['UWC', 'Max Temp (C)', 'Min Temp (C)', 'Total Precip (mm)']]

# Replace missing values with interpolation
missing('Max Temp (C)')
missing('Min Temp (C)')
missing('Total Precip (mm)')


# Split data into training/testing sets with 25% split
X_train, X_test, y_train, y_test = train_test_split(df_s, df_s['UWC'],
                                                    test_size = 0.25,
                                                    random_state = 42)

# Rename variables in train and test sets
X_train = X_train.rename(columns = {'UWC': 'SUC',
                                  'Max Temp (C)': 'Max_T',
                                  'Min Temp (C)': 'Min_T',
                                  'Total Precip (mm)':'T_P'})

X_test = X_test.rename(columns = {'UWC': 'SUC',
                                  'Max Temp (C)': 'Max_T',
                                  'Min Temp (C)': 'Min_T',
                                  'Total Precip (mm)':'T_P'})
```

## Create Model in PyMC3 and Sample from Posterior
We let a *Markov Chain Monte Carlo* algorithm draw samples from the posterior to approximate the posterior for each of the model parameters.

**Note:** Click [here](https://github.com/WillKoehrsen/DataAnalysis/blob/master/bayesian_lr/Bayesian%20Linear%20Regression%20Project.ipynb) to read more about implementing Bayesian Linear Regression using PyMC3.

```python
# Formula for Bayesian Linear Regression 
formula = 'SUC ~ ' + ' + '.join(['%s' % variable for variable in X_train.columns[1:]])
print(formula)

# Context for the model
with pm.Model() as normal_model:
    # The prior for the model parameters will be a normal distribution
    family = pm.glm.families.Normal()

    # Creating the model requires a formula and data (and optionally a family)
    pm.GLM.from_formula(formula, data = X_train, family = family)

    # Perform Markov Chain Monte Carlo sampling
    normal_trace = pm.sample(draws=2000, chains = 2, tune = 500)
```

### Trace Plot of All Samples

```python
# Define a function to draw posterior plot
def posterior_plot (variable, xlabel):
    _, ax = plt.subplots(figsize = (12, 6))
    ax.hist([n.mean() for n in normal_trace[variable]], bins = 19, alpha = 0.5 , color = 'gray', edgecolor = 'black')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set(title = 'Posterior predictive of ' + xlabel, xlabel = xlabel, ylabel = 'Frequency')

```

## Calculate Mean Absolute Error and Root Mean Square Error

```python
# Define a function to calculate MAE and RMSE
def evaluate_predictions(predictions, true):
    mae = np.mean(abs(predictions - true))
    rmse = np.sqrt(np.mean((predictions - true) ** 2))
    
    return mae, rmse

median_pred = X_train['SUC'].median()
median_preds = [median_pred for _ in range(len(X_test))]
true = X_test['SUC']

# Display mae and rmse
mae, rmse = evaluate_predictions(median_preds, true)
print('Mean Absolute Error: {:.4f}'.format(mae))
print('Root Mean Square Error: {:.4f}'.format(rmse))
```

### Plot Annual Water Consumption in 2015 - 2050

```python
# Find and generate Days of week 
df_weather['Day of Week'] = df_weather['Date'].dt.day_name()

# Find and generate Weekend
df_weather["Weekend"] = df_weather["Day of Week"].isin(["Saturday", "Sunday"])

# Assign 0.35 to water consumption on weekdays and 0.33 to consumption on weekdays (RCP 4.5 and RCP 8.5)
def bwc (tmax, tmin, UWC):
    df_weather['tmean'] = (df_weather[tmax] + df_weather[tmin])/2
    df_weather.loc[((df_weather['tmean'] < 9) & (df_weather['Weekend'] == True)), UWC] = 0.35
    df_weather.loc[((df_weather['tmean'] < 9) & (df_weather['Weekend'] == False)), UWC] = 0.33

    

bwc('tmax45', 'tmin45', 'UWC_45')
bwc('tmax85', 'tmin85', 'UWC_85')


# Convert daily water consumption to yearly values
df_y = df_weather.loc[:,['UWC_45','UWC_85','Date']]
df_y.set_index('Date', inplace = True)
df_y.index = pd.to_datetime(df_y.index)
df_yearly = df_y.resample('Y').mean()
df_yearly.reset_index(inplace = True)

# Plot anuual water consumption in 2015-2050
fig, (ax) = plt.subplots(1,1)
fig.set_size_inches(8, 6)
ax.plot(df_yearly['Date'],df_yearly['UWC_45'], color = 'black', ls = '--', label = 'RCP 4.5')
ax.plot(df_yearly['Date'],df_yearly['UWC_85'],  color = 'black', label = 'RCP 8.5')

ax.legend()
ax.set_xlabel('Date', {'fontsize': 12})
ax.set_ylabel('UWC ($m^3$/capita.day)', {'fontsize': 12})
ax.set_title('Annual urban water consumption from 2015 to 2050', {'fontsize': 15})
ax.yaxis.grid()
```
